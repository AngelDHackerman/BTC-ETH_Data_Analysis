### Objective
This notebook performs an initial exploration of the S&P 500 dataset. Key steps include:
- Loading the dataset and inspecting its structure.
- Cleaning the data by removing null values and duplicates.
- Understanding initial trends and distributions.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# Loading row data
SP_500 = pd.read_csv('./Data/raw_data_but_cleaned/sp500_data_filtered_2014.csv', encoding='cp1252')
```

Looking for general information about the dataset

```python
SP_500.info()
```

```python
SP_500.describe()
```

Checking the null values and duplicated values

```python
SP_500.isnull().sum()
```

```python
SP_500.duplicated().sum()
```

Changing Data column type to "datetime64"

```python
SP_500['Date'] = pd.to_datetime(SP_500['Date'])
```

Trim the data from a start date September 18 2014 (In order to match with the BTC dataset)

```python
# Filter Data from sep 18 2014 (same has bitcoin)
start_date = "2014-09-18"
SP_500_data_filtered = SP_500[SP_500['Date'] >= start_date]

finish_date = "2024-05-01"
SP_500_data_filtered = SP_500_data_filtered[SP_500_data_filtered['Date'] <= finish_date]
```

```python
# Convert into a CSV
SP_500_data_filtered.to_csv('./Data/raw_data_but_cleaned/sp500_data_filtered.csv', index=False)
```

Displaying grafics and initial statistics

```python
SP500_filtered = pd.read_csv('./Data/raw_data_but_cleaned/sp500_data_filtered.csv')

# Quick visualization of Bitcoin prices over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Close', data=SP500_filtered, label='S&P 500 Price', color='blue')

# Configurar ticks del eje X
plt.xticks(ticks=np.linspace(0, len(SP500_filtered) - 1, 10, dtype=int), rotation=45)  # Muestra 10 fechas espaciadas uniformemente
plt.title('S&P 500 Price Over Time (2014-2024)', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Price (USD)', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

## Observations:

- The S&P 500 dataset spans from 2014 to 2024, aligning with the Bitcoin dataset.
- No significant missing or duplicate values were found.
- The dataset shows an overall upward trend, with notable declines in [specific periods, e.g., early 2020 due to COVID-19].
- Further analysis of volatility and moving averages can reveal insights into market stability over time.
